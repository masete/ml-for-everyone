```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ['fLength','fWidth','fSize','fConc','fConc1','fAsym','fM3Long','fM3Trans','fAlpha','fDist','class']
data = pd.read_csv('data/magic04.data', names=cols)
```

```python
data.head()
```

```python
data["class"] = (data["class"] == "g").astype(int)
```

```python
data.head()
```

```python
for label in cols[:-1]:
    plt.hist(data[data['class']==1][label], color='blue',label='gamma', alpha=0.7, density=True)
    plt.hist(data[data['class']==0][label], color='red',label='hedron', alpha=0.7,density=True)
    plt.title(label)
    plt.ylabel("Probability")
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
